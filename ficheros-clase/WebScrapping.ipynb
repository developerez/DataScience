## Web Scrapping

```python
import requests
import pandas as pd
import bs4
from bs4 import BeautifulSoup
```

```python
url = "https://www.buscalibre.es/libros/search?q=Python+Web+Scrapping"
```

```python
url = requests.get(url)
```

```python
url
```

```python
cod_html =  BeautifulSoup(url.content,"html.parser")
```

```python
print(cod_html)
```

```python
nombres =cod_html.find_all("h3",{"class":"nombre"})
```

```python
print(nombres)
```

```python
print(len(nombres))
```

```python
nombre_libros = [nombres[i].get_text() for i in range (0,len(nombres))]
```
